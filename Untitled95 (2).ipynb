# Sampling out of Binomial Distribution

```python
import numpy as np
```

```python
n_defaults=np.random.binomial(n=100,p=0.05,size=10000)
```

```python
def ecdf(data):
    X=np.sort(data)
    y=np.arange(1,len(X)+1)/len(X)
    return X,y
```

```python
X,y=ecdf(n_defaults)
```

```python
import matplotlib.pyplot as plt
```

```python
_=plt.plot(X,y,linestyle="none",marker='.')
_=plt.xlabel("number of defaults out of 100 loans")
_=plt.ylabel("CDF")
plt.show()
```

```python
_=plt.hist(n_defaults, bins=100)
plt.show()
```

# Relationship between binomial and poisson distributions

```python
samples_poisson=np.random.poisson(10,size=10000)
```

```python
print("Poisson",np.mean(samples_poisson),np.std(samples_poisson))
```

```python
n=[20,100,1000]
```

```python
p=[0.5,0.1,0.01]
```

```python
for i in range(3):
    samples_binomial=np.random.binomial(n[i],p[i],size=10000)
    print("n=", n[i], "Binom:", np.mean(samples_binomial), np.std(samples_binomial))
```

# Normal Distribution

```python
samples_std1=np.random.normal(20,1,size=100000)
samples_std3=np.random.normal(20,3,size=100000)
samples_std10=np.random.normal(20,10,size=100000)
```

```python
_=plt.hist(samples_std1,bins=100,histtype="step")
_=plt.hist(samples_std3,bins=100,histtype="step")
_=plt.hist(samples_std10,bins=100,histtype="step")
plt.legend(("std=1","std=3","std=10"))
plt.show()
```

```python
x_std1,y_std1=ecdf(samples_std1)
x_std3,y_std3=ecdf(samples_std3)
x_std10,y_std10=ecdf(samples_std10)
```

```python
_=plt.plot(x_std1,y_std1,marker=".",linestyle="none")
_=plt.plot(x_std3,y_std3,marker=".",linestyle="none")
_=plt.plot(x_std10,y_std10,marker=".",linestyle="none")
plt.legend(("std=1","std=3","std=10"))
plt.show()
```
